### Part g
Create a function that takes the URL of the webpage to scrape as an input, applies the code you wrote for parts a through e, and generates the dataframe from part f as the output. [10 points]

```python
def scrape_page_for_books(url):
    headers = {'user-agent': 
        'Data Arch Pipeline class Lab 5 (ccv4xb@virginia.edu) (Language=Python 3.13.2; Platform=Mac OSX 15.6.1)'}
    r = requests.get(url, headers = headers)
    bts = BeautifulSoup(r.text, 'html.parser')
    titles =[b.attrs['title'] for b in bts.find_all('a', title=True)]
    prices = [float(p.string.replace('Â£', '')) for p in bts.find_all("p", "price_color")]
    ratings = [p['class'][1] for p in bts.find_all('p', "star-rating")]
    image_urls = [url.attrs['src'] for url in bts.find_all('img', 'thumbnail')]
    book_data = {'title': titles, 
                 'price': prices, 
                 'rating': ratings, 
                 'image_url': image_urls
                }
    books_df = pd.DataFrame(book_data)
    return books_df
```
